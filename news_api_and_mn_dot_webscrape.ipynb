```python
news_all_df = news_all_df.dropna(subset=['content'])
```

```python
news_all_df.content.isnull().sum()
```

```python
locations = []
for i in news_all_df['content']:
    doc = nlp(i)
    #pprint([(X.text, X.label_) for X in doc.ents])
    locations.append([(X.text, X.label_) for X in doc.ents])
```

```python
len(locations)
```

```python
# the GPE list contains more categories than we are interested in.  we really are only looking for locations and dates
#so we will filter the list here to those two categories before we append it to our dataframe
locations_and_dates = []

for i in locations:
    #print(i)
    locs = []
    for j in i:
        #print(j)
        if j[1] in ('DATE', 'GPE'):
            #print(j)
            locs.append(j)
    locations_and_dates.append(locs)
        
   # for j in i:
     #   print(j)
        
   # print(i)
```

```python
news_all_df['locations_and_times'] = locations_and_dates
```
